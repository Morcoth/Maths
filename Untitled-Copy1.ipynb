1. Задание
Даны два вектора в трехмерном пространстве: (20,20,20) и (0,0,-20)
Найдите их сумму. (на листочке)
Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

x=np.linspace(0,20,100)
k=1
yx1 = np.cos(k*x)
k=0.5
yx2 = np.cos(k*x)
plt.plot(x, yx1)
plt.plot(x, yx2)
plt.show()
```

```python
import matplotlib.pyplot as plt
circle1=plt.Circle((0.6,0.6),0.2,color='r')
plt.gcf().gca().add_artist(circle1)
```

```python
from matplotlib.patches import Ellipse
e1 = Ellipse((0.5,0.5), 0.2, 0.4)
fig, ax = plt.subplots(subplot_kw={'aspect': 'equal'})
ax.add_artist(e1)
plt.show()
```

```python
x = y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
a = 1
b = 2
plt.contour(x, y,(x**2/a**2 - y**2/b**2), [1], colors='k')
plt.show()
```

Задание (в программе)
Нарисуйте трехмерный график двух параллельных плоскостей.
Нарисуйте трехмерный график двух любых поверхностей второго порядка.

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D

fig = figure()
ax=Axes3D(fig)
X = Y = np.arange(-5,5,0.5)
X1 = Y1 = np.arange(-5,5,1)
X1, Y1 = np.meshgrid(X1, Y1)
X, Y = np.meshgrid(X,Y)
Z=2*X+5*Y
Z1 = 2*X1 + 5*Y1 + 20
ax.plot_wireframe (X,Y,Z)
ax.plot_wireframe (X1,Y1,Z1)
show()
```
